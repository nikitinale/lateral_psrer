```python
from shapely.geometry import box

def fishnet(geometry, threshold):
  '''
  divide area on small squares for representation of contamination level
  '''
  bounds = geometry.bounds
  xmin = int(bounds[0] // threshold)
  xmax = int(bounds[2] // threshold)
  ymin = int(bounds[1] // threshold)
  ymax = int(bounds[3] // threshold)
  ncols = int(xmax - xmin + 1)
  nrows = int(ymax - ymin + 1)
  result = []
  for i in range(xmin, xmax+1):
    for j in range(ymin, ymax+1):
      b = box(i*threshold, j*threshold, (i+1)*threshold, (j+1)*threshold)
      g = geometry.intersection(b)
      if g.is_empty:
        continue
      result.append(g)
  return result
```

```python
from shapely.geometry import Point, Polygon
import geopandas as gpd
# бибилиотека отображения гео-информации на карте
import geoplot as gplt
# импорт геопроекции для отображения информации на картах
import geoplot.crs as gcrs

idx = pd.IndexSlice

import fiona
fiona.drvsupport.supported_drivers['KML'] = 'rw' # enable KML support which is disabled by default

wsfile = 'Perstok_catchment.kml'
ws = gpd.read_file(wsfile)
ws['geometry'] = ws['geometry'].to_crs(epsg=28406)
ws = ws.set_index('Name')
contour = ws.loc['Catchment', 'geometry'] - ws.loc['Perstok', 'geometry']
threshold = 20
net = fishnet(contour, threshold)

# df_contamination ['year', 'x', 'y', 'cs137_top_kbq_sqm', 'cs137_bottom_kbq_sqm', 'am241_top_kbq_sqm', 'am241_bottom_kbq_sqm']
# sum of the radioisotopes stack in all layers of soil
df_contamination['Cs-137'] = df_contamination['cs137_top_kbq_sqm'] + df_contamination['cs137_bottom_kbq_sqm']
df_contamination['Am-241'] = df_contamination['am241_top_kbq_sqm'] + df_contamination['am241_bottom_kbq_sqm']

df_contamination = pd.merge(df_contamination, coordinates, left_on=['x', 'y'], right_on=['x', 'y'], how='inner')
# random distribution of points around initial coordinates for smoothing result
df_contamination.xg = df_contamination.xg + np.random.randn(df_contamination.xg.shape[0])*35
df_contamination.yg = df_contamination.yg + np.random.randn(df_contamination.yg.shape[0])*35
geometry = [Point(xy) for xy in zip(df_contamination.xg, df_contamination.yg)]
gdf_contamination = gpd.GeoDataFrame(df_contamination, crs={'init': 'epsg:28406'}, geometry=geometry)
gdf_contamination.geometry = gdf_contamination.geometry.buffer(20)

gdf_contamination['geometry'] = gdf_contamination['geometry'].to_crs(epsg=4326)
```

```python
netdf = gpd.GeoDataFrame(geometry=net, crs={'init': 'epsg:28406'})
netdf['geometry'] = netdf['geometry'].to_crs(epsg=4326)
```

```python
from shapely.wkt import loads

def plot_map(year, isotop):
  # select by year and only land surface
  tomap = gdf_contamination[(df_contamination['year'] == year) & (df_contamination['land_type'] != 'w')]
  # spatial join of generated in Monte-Karlo values of contamination with squares divides territory of cathcment
  mapa = gpd.sjoin(netdf, tomap, how='inner', op='intersects')
  mapa['WKT'] = mapa['geometry'].apply(lambda x: str(x))
  gr_mapa = mapa.groupby('WKT').agg(['mean']).reset_index() # mean level of contamination in the squares
  gr_mapa['geometry'] = gr_mapa['WKT'].apply(lambda x: loads(x))
  gr_mapa = gpd.GeoDataFrame(gr_mapa)
  mapa = gpd.GeoDataFrame(mapa, geometry=mapa.geometry, crs={'init': 'epsg:4326'})
  # plot mean level of contamination for squares with isotop
  mapa.plot(column=isotop, legend=True, cmap='Reds', figsize=(10,10)) #, vmin=0, vmax=3600)
  plt.title(f'Contamination of catchment area of the lake Perstok with {isotop} at {year} year, kBq/sq.m.')

plot_map(2019, 'Cs-137')
```
